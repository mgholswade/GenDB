```python
f = open("test.txt", "r")
data = [line.split() for line in f]

data = data[1:]
masterList = []
for line in data:
    newlist = []
    for item in line:
        newlist.append(item)
    masterList.append(newlist)

df = pd.DataFrame.from_dict(masterList)
df.columns = columns=['Scaffold','Source','Type','Start','End','A','B','C','Name','D','E','F','G','H']

```
